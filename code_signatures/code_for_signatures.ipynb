Supplementary Table 1. Cellular Senescence Markers Database. This table containing all the senescent markers described in our literature search. We organized working groups for each section of the manuscript, aiming for comprehensive coverage within the constraints of the manuscript's format. Specifically, for the tissue sections, we implemented consistent inclusion criteria for published studies: 1) evidence of senescence in mouse or human tissues, 2) description of normal aging or disease states, and 3) provision of two or more hallmarks of senescence. As part of this process, each group collected the list of markers that were described in the manuscripts that fulfilled the selection criteria and compiled, which were then combined into this supplemental table. For each marker the following information was collected: 1) the marker name or symbol, 2) its full name or preferred term, 3) the marker HGNC ID, 4) the biomolecule type , 5) the tissue where the marker was described, 6) Cell type where it was detected, 7) the Context in which it was studied, 8) the Organism (Mouse/Human), 9) the Senescence Hallmark the marker corresponds to, 10) the PUBMED ID of the publications where the marker was reported, 11) the techniques used to detect the marker. The list of abbreviations can be found in the caption for Table 1. An interactive visual representation of this database can be accessed at https://docs.sennetconsortium.org/biomarkers/.

```python
df4 = df4[['Symbol', 'Type', 'Functional category', 'Sub1-category']] \
         .rename(columns={'Symbol': 'Gene'})
```

```python
mg = mygene.MyGeneInfo()
gene_info = mg.querymany(df4['Gene'], scopes='symbol', fields='ensembl.gene', species='human')

gene_to_ensembl = {}

for gene in gene_info:
    if 'ensembl' in gene:
        if isinstance(gene['ensembl'], list): 
            gene_to_ensembl[gene['query']] = gene['ensembl'][0]['gene']
        else: 
            gene_to_ensembl[gene['query']] = gene['ensembl']['gene']
    else:
        gene_to_ensembl[gene['query']] = None  

df4['ENSEMBL'] = df4['Gene'].map(gene_to_ensembl)
```

```python

missing_genes_ensembl = df4.merge(df24[['ENSEMBL']], 
                                on='ENSEMBL', 
                                how='left', 
                                indicator=True) \
                        .query('_merge == "left_only"')['ENSEMBL'] \
                        .unique()


missing_genes_names = df4[df4['ENSEMBL'].isin(missing_genes_ensembl)]['Gene'].unique()

missing_genes_names
```

```python

def is_gene_in_synonyms(gene, synonyms_str):
    if pd.isna(synonyms_str):
        return False
    synonyms = [s.strip() for s in str(synonyms_str).split(',')]
    return gene in synonyms


double_check_results = []
for gene in missing_genes_names:
    in_main = gene in df24['Gene'].values
    in_synonyms = any(df24['Gene Synonyms'].apply(lambda x: is_gene_in_synonyms(gene, x)))
    
    if not in_main and not in_synonyms:
        ensembl_id = df4[df4['Gene'] == gene]['ENSEMBL'].values[0]
        double_check_results.append({'ENSEMBL': ensembl_id, 'Gene': gene})


missing_genes_double_check = pd.DataFrame(double_check_results)
missing_genes_double_check['Status'] = 'Not found in Gene or Gene Synonyms'

print(f"Found {len(missing_genes_double_check)} genes missing in both columns:")
print(missing_genes_double_check)
```

```python

genes_to_add = missing_genes_double_check.merge(
    df4[['Gene', 'ENSEMBL', 'Type']], 
    on=['Gene', 'ENSEMBL'],
    how='left'
)

genes_to_add['Class'] = 'Chaperone'
genes_to_add['Hallmark'] = 'Proteostasis'
genes_to_add['Subprocess'] = 'Cytonuclear proteostasis'
genes_to_add['Gene Synonyms'] = 'Unknown'

genes_to_add = genes_to_add.rename(columns={'Type': 'Group'})

cols_in_df24 = df24.columns.tolist()
genes_to_add = genes_to_add[cols_in_df24]
df24_updated = pd.concat([df24, genes_to_add], ignore_index=True)

print(f"Added {len(genes_to_add)} new genes:")
print(genes_to_add[['Gene', 'ENSEMBL', 'Group']].head())
```

```python
def merge_gene_data(gene_corr_df, all_hallmarks, species='Mouse', experiment_name='experiment'):
    """
    Merges gene correlation data with hallmark annotations.

    Parameters:
    - gene_corr_df: DataFrame with correlation data (must contain 'Gene', 'SpearmanR', 'p_value')
    - all_hallmarks: DataFrame with annotations (must contain 'Gene', 'Subprocess', 'Hallmark', 'ENSEMBL', 'ENSEMBL_mice')
    - species: 'Mouse' or 'Human' (default 'Mouse')
    - experiment_name: experiment name for the new column

    Returns:
    - Merged DataFrame in the required structure
    """

    id_col = 'ENSEMBL_mice' if species == 'Mouse' else 'ENSEMBL'
    gene_corr_renamed = gene_corr_df.rename(columns={'Gene': id_col})
    
    merged = pd.merge(
        all_hallmarks,
        gene_corr_renamed,
        on=id_col,
        how='left'
    )
    
    merged['Experiment'] = experiment_name
    merged['Species'] = species
    
    result = merged[[
        'Hallmark',
        'Subprocess',
        'Gene',
        'SpearmanR',
        'p_value',
        'ENSEMBL',
        'ENSEMBL_mice',
        'Experiment',
        'Species'
        
    ]]
    result = result.rename(columns={'Gene': 'Gene Symbol'})
    result = result.rename(columns={'SpearmanR': 'Spearman'})
    return result

# example of usage :
# result_df = merge_gene_data(
#     gene_corr_df,
#     all_hallmarks,
#     species='Mouse',
#     experiment_name='reprogramming_TOR'
# )
```

```python
def count_significant_genes(result_df):
    """
    Counts genes passing correlation thresholds (0.2 and 0.4)
    for each Subprocess-Hallmark combination.

    Parameters:
    - result_df: Results DataFrame (must contain 'Subprocess', 'Hallmark',
                'Spearman', 'Experiment', 'Species')

    Returns:
    - DataFrame with gene counts for each threshold
    """
    grouped = result_df.groupby(['Subprocess', 'Hallmark', 'Experiment', 'Species'])
    
    result = grouped.agg(
        n_genes_ge_0_2=('Spearman', lambda x: (x >= 0.2).sum()),
        n_genes_le_minus_0_2=('Spearman', lambda x: (x <= -0.2).sum()),
        n_genes_ge_0_4=('Spearman', lambda x: (x >= 0.4).sum()),
        n_genes_le_minus_0_4=('Spearman', lambda x: (x <= -0.4).sum()),
        total_genes=('Spearman', 'count')
    ).reset_index()

    result = result.rename(columns={'n_genes_ge_0_2': 'n genes >= 0.2'})
    result = result.rename(columns={'n_genes_le_minus_0_2': 'n genes <= 0.2'})
    result = result.rename(columns={'n_genes_ge_0_4': 'n genes >= 0.4'})
    result = result.rename(columns={'n_genes_le_minus_0_4': 'n genes <= 0.4'})

    result = result[[
        'Hallmark',
        'Subprocess',
        'n genes >= 0.2',
        'n genes <= 0.2',
        'n genes >= 0.4',
        'n genes <= 0.4',
        'Experiment',
        'Species'

    ]]
    
    return result

# gene_counts_df = count_significant_genes(result_df)
# gene_counts_df
```
